```python
import numpy as np
import matplotlib.pyplot as plt

# Define vectors
# v1 = np.array([1, 2])  # Base vector
# v2 = np.array([2, 3])  # First comparison vector
# v3 = np.array([-1, -2])  # Second comparison vector

np.random.seed()  # Ensure randomness
v1 = np.random.uniform(-3, 3, size=2)  # Base vector
v2 = np.random.uniform(-3, 3, size=2)  # First comparison vector
v3 = np.random.uniform(-3, 3, size=2)  # Second comparison vector

# Compute dot products
dot_v1_v2 = np.dot(v1, v2)
dot_v1_v3 = np.dot(v1, v3)

# Compute cosine similarities
cos_sim_v1_v2 = dot_v1_v2 / (np.linalg.norm(v1) * np.linalg.norm(v2))
cos_sim_v1_v3 = dot_v1_v3 / (np.linalg.norm(v1) * np.linalg.norm(v3))

# Print similarities
print(f"Dot product (v1, v2): {dot_v1_v2}")
print(f"Dot product (v1, v3): {dot_v1_v3}")
print(f"Cosine similarity (v1, v2): {cos_sim_v1_v2}")
print(f"Cosine similarity (v1, v3): {cos_sim_v1_v3}")

# Plot the vectors
origin = [0], [0]  # Origin for the vectors
plt.quiver(*origin, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='r', label='v1')
plt.quiver(*origin, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='g', label='v2')
plt.quiver(*origin, v3[0], v3[1], angles='xy', scale_units='xy', scale=1, color='b', label='v3')

# Add legend and grid
plt.legend()
plt.grid()
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Vector Similarity')
plt.show()
```
