```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from scipy.linalg import expm, eigh  # Use scipy for the matrix exponential function

class Simulator:
    def __init__(self, N, H):
        self.N = N
        self.H = H
        
    def classical(self):
        """Classically find the optimal configuration and energy in second quantization."""
        best_energy = float('inf')
        best_config = None

        # Brute-force over all occupation configurations
        for config in range(1 << self.N):
            # Convert config index to occupation numbers
            occupations = [(config >> i) & 1 for i in range(self.N)]

            # Create the corresponding state vector
            state_vector = np.zeros(1 << self.N, dtype=complex)
            state_vector[config] = 1.0

            # Calculate the energy for this configuration
            energy = np.vdot(state_vector, self.H @ state_vector).real

            # Debugging output
            print(f"Config {config}: Occupations = {occupations}, Energy = {energy}")

            # Check if this is the best configuration
            if energy < best_energy:
                best_energy = energy
                best_config = occupations

        return best_config, best_energy
    
    def creation_operator(self, state, site):
        new_state = state.copy()
        if new_state[site] == 1:  # Fermions: max occupation is 1
            return None, 0  # Invalid operation
        new_state[site] += 1
        return new_state, np.sqrt(1)

    def annihilation_operator(self, state, site):
        new_state = state.copy()
        if new_state[site] == 0:
            return None, 0  # Invalid operation
        new_state[site] -= 1
        return new_state, np.sqrt(1)

    def hamiltonian_operator(self, state, eps, J):
        N = len(state)
        H_matrix = np.zeros((2**N, 2**N), dtype=complex)

        for config in range(1 << N):
            spins = [(config >> k) & 1 for k in range(N)]

            # On-site term
            for i in range(N):
                H_matrix[config, config] += eps[i] * spins[i]

            # Hopping term
            for i in range(N):
                for j in range(N):
                    if i != j:
                        new_spins, coeff = self.creation_operator(spins, i)
                        if new_spins:
                            new_spins, coeff2 = self.annihilation_operator(new_spins, j)
                            if new_spins:
                                new_config = sum((1 << k) * new_spins[k] for k in range(N))
                                H_matrix[config, new_config] += J[i][j] * coeff * coeff2
        return H_matrix
    
    def compute_eigenvalues(self):
        """Compute all eigenvalues of the Hamiltonian matrix."""
        if isinstance(self.H, np.ndarray):  # Ensure H is a matrix
            eigenvalues = np.linalg.eigvalsh(self.H)  # Use Hermitian eigenvalue solver
            return np.sort(eigenvalues)
        else:
            raise ValueError("Hamiltonian is not a valid matrix.")
            
    def suggest_t(self):
        """Suggest an optimal time t based on the energy gap."""
        eigenvalues = self.compute_eigenvalues()
        unique_eigenvalues = np.unique(eigenvalues)

        if len(unique_eigenvalues) < 2:
            raise ValueError("Insufficient unique eigenvalues to compute energy gap.")

        gaps = np.diff(unique_eigenvalues)  # Compute energy gaps
        delta_E = np.min(gaps)

        #if delta_E <= 1e-10:  # Tolerance for small gaps
        #    raise ValueError("Energy gap is too small or non-positive.")

        return self.N / (delta_E ** 2)  # Suggest t proportional to the inverse gap


    def evolve(self, num_steps=100, T=10):
        N = self.N
        num_states = 2**N
        eps = np.random.rand(N)  # Example on-site energies
        J = np.random.rand(N, N)  # Example hopping matrix

        H_initial = self.hamiltonian_operator([0] * N, eps, J)
        H_final = self.H  # Assume user provides in second quantized form

        time_points = np.linspace(0, T, num_steps)
        #current_state = np.ones(num_states, dtype=complex) / np.sqrt(num_states)
        eigvals, eigvecs = np.linalg.eigh(H_initial)
        current_state = eigvecs[:, np.argmin(eigvals)]
        
        states_over_time = []
        H_ts_over_time = []

        states_over_time = []
        for t in time_points:
            s = t / T
            H_t = (1 - s) * H_initial + s * H_final
            U = expm(-1j * H_t * (T / num_steps))
            current_state = U @ current_state
            states_over_time.append((t, current_state))

        return states_over_time, H_ts_over_time, current_state

```

```python
def measure(final_state, simulator, shots=10000):
    """
    Measure the final quantum state and return the most frequently observed configuration and its energy.
    :param final_state: The final quantum state as a complex vector.
    :param simulator: The simulator instance with Hamiltonian matrix H.
    :param shots: Number of measurements to simulate.
    :return: The most frequently observed configuration and its energy.
    """
    # Compute probabilities for each configuration
    probabilities = np.abs(final_state)**2
    probabilities /= np.sum(probabilities)  # Normalize

    # Perform sampling based on probabilities
    sampled_indices = np.random.choice(len(probabilities), size=shots, p=probabilities)

    # Count the occurrences of each index
    unique, counts = np.unique(sampled_indices, return_counts=True)
    most_frequent_index = unique[np.argmax(counts)]

    # Convert the most frequent index to an occupation number configuration
    occupations = [(most_frequent_index >> i) & 1 for i in range(simulator.N)]

    # Calculate the energy for the most frequently observed configuration
    # Assuming simulator.H is a Hamiltonian matrix
    sampled_state = np.zeros_like(final_state)
    sampled_state[most_frequent_index] = 1.0
    energy = np.vdot(sampled_state, simulator.H @ sampled_state).real

    # Debugging: Print details of the measurement
    print(f"Most Frequent Index: {most_frequent_index}")
    print(f"Most Frequent Occupations: {occupations}")
    print(f"Most Frequent Energy: {energy}")

    return occupations, energy
```

```python
class HaldaneShastrySecondQuant:
    def __init__(self, vertices, J0):
        """
        Initialize the Haldane-Shastry model in second quantization.
        :param vertices: List of vertices (sites).
        :param J0: Coupling constant J_0.
        """
        self.vertices = vertices
        self.num_vertices = len(vertices)
        self.J0 = J0
        self.edges = [(i, j) for i in range(self.num_vertices)
                      for j in range(i + 1, self.num_vertices)]
        self.weights = {(i, j): J0 / (j - i) ** 2 for i, j in self.edges}

    def hamiltonian(self):
        """Define the Hamiltonian matrix in the Fock basis."""
        N = self.num_vertices
        H_matrix = np.zeros((2**N, 2**N), dtype=complex)
        
        for config in range(1 << N):
            spins = [(config >> k) & 1 for k in range(N)]  # Fock state as a binary list
            
            for i, j in self.edges:
                # Translate spin interaction into Fock basis
                n_i = spins[i]
                n_j = spins[j]
                S_i = 2 * n_i - 1  # Map Fock state to spin state
                S_j = 2 * n_j - 1
                H_matrix[config, config] += self.weights[(i, j)] * S_i * S_j
        
        return H_matrix
```

```python
# Define vertices and coupling constant J0
vertices = [0, 1, 2, 3]
J0 = 1.0

# Create the Haldane-Shastry model in second quantization
hs_second_quant = HaldaneShastrySecondQuant(vertices, J0)
H_second_quant = hs_second_quant.hamiltonian()

# Perform Adiabatic Evolution
sim = Simulator(len(vertices), H_second_quant)

# Classical Simulation
best_config, best_energy = sim.classical()
print(f"Classical Best Occupations: {best_config}")
print(f"Classical Best Energy: {best_energy}")


# Suggested optimal t (if implemented in your Simulator class)
optimal_t = sim.suggest_t()
print(f"Suggested optimal t: {optimal_t}")

# Adiabatic evolution
states_forward, H_ts_over_time, final_state = sim.evolve(num_steps=200, T=optimal_t)

# Measure the final state
best_config, best_energy = measure(final_state, sim)
print("Final State Measurement:")
print(f"Best Occupations: {best_config}")
print(f"Best Energy: {best_energy}")
```

```python
def print_top_occupation_probabilities(final_state, simulator, top_n=5):
    """
    Print the top N occupation state probabilities from the final quantum state.
    :param final_state: The final quantum state as a complex vector.
    :param simulator: The simulator instance for energy calculation.
    :param top_n: Number of top probabilities to print.
    """
    # Compute probabilities for each configuration
    probabilities = np.abs(final_state)**2
    probabilities /= np.sum(probabilities)  # Normalize

    # Get indices sorted by probabilities in descending order
    sorted_indices = np.argsort(probabilities)[::-1]

    # Print the top N configurations
    print(f"Top {top_n} occupation state probabilities:")
    for rank in range(top_n):
        index = sorted_indices[rank]
        prob = probabilities[index]
        
        # Convert the index to an occupation configuration
        occupations = [(index >> i) & 1 for i in range(simulator.N)]

        # Create a state vector for energy calculation
        state_vector = np.zeros_like(final_state)
        state_vector[index] = 1.0
        
        # Calculate energy using the Hamiltonian matrix
        energy = np.vdot(state_vector, simulator.H @ state_vector).real

        print(f"Rank {rank + 1}: Occupations = {occupations}, Probability = {prob:.4f}, Energy = {energy:.4f}")

print_top_occupation_probabilities(final_state, sim, top_n=5)
```

```python
def plot_state_probabilities(states_over_time, classical_best=None):
    """
    Plots the evolution of state probabilities over time for the second quantization simulator.

    Parameters:
        states_over_time (list): List of tuples (time, state vector) from adiabatic evolution.
        classical_best (list): Classical best occupation (optional, to highlight its state).
    """
    num_states = len(states_over_time[0][1])
    probabilities_over_time = []
    time_points = []

    for t, state in states_over_time:
        probabilities = np.abs(state)**2
        probabilities_over_time.append(probabilities)
        time_points.append(t)

    probabilities_over_time = np.array(probabilities_over_time)
    time_points = np.array(time_points)

    # Plot state probabilities
    plt.figure(figsize=(12, 6))
    colors = cm.viridis(np.linspace(0, 1, num_states))

    for state_idx in range(num_states):
        plt.plot(
            time_points,
            probabilities_over_time[:, state_idx],
            color=colors[state_idx],
            alpha=0.8,
            label=f"State {state_idx}" if state_idx < 5 else None,  # Label first 5 states
        )

    # Highlight classical best occupation (if provided)
    if classical_best is not None:
        classical_index = sum((1 << i) * b for i, b in enumerate(classical_best))
        plt.plot(
            time_points,
            probabilities_over_time[:, classical_index],
            color="red",
            alpha=1.0,
            linewidth=2,
            label=f"Classical Best: {classical_best}",
        )

    plt.title("State Probabilities Over Time")
    plt.xlabel("Time")
    plt.ylabel("Probability")
    plt.grid(True)
    plt.legend(loc="upper right", fontsize="small")
    plt.tight_layout()
    plt.show()

plot_state_probabilities(states_forward)
```

```python
class XXZModelSecondQuant:
    def __init__(self, vertices, Jx, Jy, Jz, alpha=2.0):
        """
        Initialize the XXZ Model in second quantization.
        :param vertices: List of vertices (spin sites).
        :param Jx: Coupling constant for the x direction.
        :param Jy: Coupling constant for the y direction.
        :param Jz: Coupling constant for the z direction.
        :param alpha: Decay exponent for long-range interactions.
        """
        self.vertices = vertices
        self.num_vertices = len(vertices)
        self.edges = [(i, j) for i in range(self.num_vertices) for j in range(i + 1, self.num_vertices)]
        self.Jx = Jx
        self.Jy = Jy
        self.Jz = Jz
        self.alpha = alpha
        self.weights = {
            (i, j): 1.0 / abs(i - j) ** self.alpha
            for i, j in self.edges
        }

    def hamiltonian(self):
        """Define the XXZ Hamiltonian matrix in the Fock basis."""
        N = self.num_vertices
        H_matrix = np.zeros((2**N, 2**N), dtype=complex)

        for config in range(1 << N):
            spins = [(config >> k) & 1 for k in range(N)]  # Fock state as binary list

            # On-site terms for σ^z
            for (i, j), weight in self.weights.items():
                S_i = 2 * spins[i] - 1  # Map occupation to spin state
                S_j = 2 * spins[j] - 1
                H_matrix[config, config] += self.Jz * weight * S_i * S_j

            # Off-diagonal terms for σ^x and σ^y
            for (i, j), weight in self.weights.items():
                # σ^x σ^x term
                new_spins, coeff = self.flip_spin(spins, i, j)
                if new_spins:
                    new_config = sum((1 << k) * new_spins[k] for k in range(N))
                    H_matrix[config, new_config] += self.Jx * weight * coeff

                # σ^y σ^y term
                new_spins, coeff = self.flip_spin(spins, i, j, include_phase=True)
                if new_spins:
                    new_config = sum((1 << k) * new_spins[k] for k in range(N))
                    H_matrix[config, new_config] += self.Jy * weight * coeff

        return H_matrix

    @staticmethod
    def flip_spin(state, i, j, include_phase=False):
        """
        Flip spins at sites i and j and return the new state and phase.
        :param state: Current occupation state.
        :param i: Site i.
        :param j: Site j.
        :param include_phase: Whether to include a phase factor for σ^y.
        :return: New state and coefficient (phase factor).
        """
        new_state = state.copy()
        if new_state[i] == new_state[j]:  # Flip not allowed if spins are the same
            return None, 0

        # Flip spins at i and j
        new_state[i] = 1 - new_state[i]
        new_state[j] = 1 - new_state[j]

        phase = (-1j) if include_phase else 1.0
        return new_state, phase
```

```python
def print_top_fock_state_probabilities(final_state, simulator, top_n=5):
    """
    Print the top N Fock state probabilities from the final quantum state.
    :param final_state: The final quantum state as a complex vector.
    :param simulator: The simulator instance for energy calculation.
    :param top_n: Number of top probabilities to print.
    """
    # Compute probabilities for each configuration
    probabilities = np.abs(final_state)**2
    probabilities /= np.sum(probabilities)  # Normalize
    
    print(f"Probabilities: {probabilities}")

    # Get indices sorted by probabilities in descending order
    sorted_indices = np.argsort(probabilities)[::-1]

    # Print the top N configurations
    print(f"Top {top_n} Fock state probabilities:")
    for rank in range(top_n):
        index = sorted_indices[rank]
        prob = probabilities[index]

        # Convert the index to occupation number configuration
        occupations = [(index >> i) & 1 for i in range(simulator.N)]

        # Calculate the energy for this configuration
        state_vector = np.zeros_like(final_state)
        state_vector[index] = 1.0  # Single-entry vector for the current configuration
        energy = np.vdot(state_vector, simulator.H @ state_vector).real

        print(f"Rank {rank + 1}: Occupations = {occupations}, Probability = {prob:.4f}, Energy = {energy:.4f}")
        
print_top_fock_state_probabilities(final_state, sim)
```
