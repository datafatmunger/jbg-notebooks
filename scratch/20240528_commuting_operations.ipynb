```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the rotation matrices
def rotation_x(theta):
    """
    Rotation matrix around the x-axis by theta radians
    """
    return np.array([
        [1, 0, 0],
        [0, np.cos(theta), -np.sin(theta)],
        [0, np.sin(theta), np.cos(theta)]
    ])

def rotation_y(theta):
    """
    Rotation matrix around the y-axis by theta radians
    """
    return np.array([
        [np.cos(theta), 0, np.sin(theta)],
        [0, 1, 0],
        [-np.sin(theta), 0, np.cos(theta)]
    ])

# Rotation angles
angle_90 = np.pi/2  # 90 degrees
angle_45 = np.pi/4  # 45 degrees

# Combine rotation matric
rotation_matrix = np.dot(rotation_x(angle_45), rotation_x(angle_90))

# Define a cube in 3D space
points = np.array([
    [0, 0, 0],  # Vertex 1
    [1, 0, 0],  # Vertex 2
    [1, 1, 0],  # Vertex 3
    [0, 1, 0],  # Vertex 4
    [0, 0, 1],  # Vertex 5
    [1, 0, 1],  # Vertex 6
    [1, 1, 1],  # Vertex 7
    [0, 1, 1]   # Vertex 8
])

# Apply rotation
rotated_points = np.dot(points, rotation_matrix)

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot original cube
ax.scatter(points[:,0], points[:,1], points[:,2], color='b')
for i in range(4):
    ax.plot([points[i,0], points[(i+1)%4,0]], 
            [points[i,1], points[(i+1)%4,1]], 
            [points[i,2], points[(i+1)%4,2]], color='b')
ax.plot([points[0,0], points[4,0]], [points[0,1], points[4,1]], [points[0,2], points[4,2]], color='b')
ax.plot([points[1,0], points[5,0]], [points[1,1], points[5,1]], [points[1,2], points[5,2]], color='b')
ax.plot([points[2,0], points[6,0]], [points[2,1], points[6,1]], [points[2,2], points[6,2]], color='b')
ax.plot([points[3,0], points[7,0]], [points[3,1], points[7,1]], [points[3,2], points[7,2]], color='b')

# Plot rotated cube
ax.scatter(rotated_points[:,0], rotated_points[:,1], rotated_points[:,2], color='r')
for i in range(4):
    ax.plot([rotated_points[i,0], rotated_points[(i+1)%4,0]], 
            [rotated_points[i,1], rotated_points[(i+1)%4,1]], 
            [rotated_points[i,2], rotated_points[(i+1)%4,2]], color='r')
ax.plot([rotated_points[0,0], rotated_points[4,0]], [rotated_points[0,1], rotated_points[4,1]], [rotated_points[0,2], rotated_points[4,2]], color='r')
ax.plot([rotated_points[1,0], rotated_points[5,0]], [rotated_points[1,1], rotated_points[5,1]], [rotated_points[1,2], rotated_points[5,2]], color='r')
ax.plot([rotated_points[2,0], rotated_points[6,0]], [rotated_points[2,1], rotated_points[6,1]], [rotated_points[2,2], rotated_points[6,2]], color='r')
ax.plot([rotated_points[3,0], rotated_points[7,0]], [rotated_points[3,1], rotated_points[7,1]], [rotated_points[3,2], rotated_points[7,2]], color='r')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
