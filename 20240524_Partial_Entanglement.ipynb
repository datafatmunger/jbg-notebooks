# Partial Entanglement

```python
%matplotlib inline
from qiskit import QuantumCircuit, transpile
from qiskit.providers.basic_provider import BasicSimulator
from qiskit.visualization import plot_histogram

from math import asin, sqrt
```

```python
# Partial Entanglement

# Create a Quantum Circuit with 2 qubits
qc = QuantumCircuit(2, 2)

# Apply Hadamard gate to the first qubit to create superposition
qc.h(0)

# Apply a controlled-Ry gate to the second qubit, controlled by the first qubit
theta = 2 * asin(1 / sqrt(3))
qc.cry(theta, 0, 1)

qc.measure(0, 0)
qc.measure(1, 1)


```

```python
# Specify the backend (the simulator)
simulator = BasicSimulator()

# Transpile the circuit for the simulator
compiled_circuit = transpile(qc, simulator)

# Run the compiled circuit
job = simulator.run(compiled_circuit)

# Get the result
result = job.result()

# Print the counts
print(result.get_counts())

# Plot the histogram
plot_histogram(result.get_counts())
```
