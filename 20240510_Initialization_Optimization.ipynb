```python
# Add your code below
#################################################
from qiskit.quantum_info import partial_trace, Statevector
from qiskit.visualization import plot_bloch_multivector

# Create a quantum register of 3 qubits
qreg = QuantumRegister(3)

# create a classical register of 3 bits
creg = ClassicalRegister(3)

# create a quantum circuit
qc = QuantumCircuit(qreg, creg)

# apply hadamard gate to the first qubit
qc.h(qreg[0])

# Apply X gate to the second qubit
qc.x(qreg[1])

# Apply Z gate to the third qubit
qc.z(qreg[2])

# draw the circuit
#qc.draw(output='mpl')

# measure the qubit
qc.measure(qreg[0], creg[0])
qc.measure(qreg[1], creg[1])
qc.measure(qreg[2], creg[2])

# Specify the backend (the simulator)
simulator = BasicSimulator()

# Transpile the circuit for the simulator
compiled_circuit = transpile(qc, simulator)

# Run the compiled circuit
job = simulator.run(compiled_circuit)

# Get the result
result = job.result()

# Print the counts
print(result.get_counts())

# Plot the histogram
plot_histogram(result.get_counts())

#################################################
```

```python
# Add your code below
#################################################
from qiskit.quantum_info import partial_trace, Statevector
from qiskit.visualization import plot_bloch_multivector

# Create a quantum register of 3 qubits
qr = QuantumRegister(4)

# create a classical register of 3 bits
cr = ClassicalRegister(4)

# create a quantum circuit
qc = QuantumCircuit(qr, cr)

qc.h(qr[0])
qc.cx(qr[0], qr[1])
qc.cx(qr[1], qr[2])
qc.cx(qr[2], qr[3])

qc.x(qr[1])
qc.x(qr[3])


# draw the circuit
qc.draw(output='mpl')

```

```python
# measure the qubit
qc.measure(qr[0], cr[0])
qc.measure(qr[1], cr[1])
qc.measure(qr[2], cr[2])
qc.measure(qr[3], cr[3])

# Specify the backend (the simulator)
simulator = BasicSimulator()

# Transpile the circuit for the simulator
cc = transpile(qc, simulator)

# Run the compiled circuit
job = simulator.run(cc)

# Get the result
result = job.result()

# Print the counts
print(result.get_counts())

# Plot the histogram
plot_histogram(result.get_counts())

#################################################
```
