```python
import pennylane as qml
from pennylane import numpy as np
import matplotlib.pyplot as plt

# Setup
num_points = 8  # For example
num_qubits = 3  # Enough to index 8 points
critical_radius = 0.75  # Adjustable
dev = qml.device("default.qubit", wires=num_qubits)

# Fake 3D dataset
np.random.seed(42)
points = np.random.rand(num_points, 3)
dist_matrix = np.linalg.norm(points[:, None, :] - points[None, :, :], axis=2)
```

```python
@qml.qnode(dev)
def local_density_circuit(index):
    bin_state = np.array(list(np.binary_repr(index, width=num_qubits)), dtype=int)

    qml.BasisState(bin_state, wires=range(num_qubits))
    for wire in range(num_qubits):
        qml.Hadamard(wires=wire)

    # Oracle flip for neighbors within critical radius
    for i in range(num_points):
        if dist_matrix[index, i] < critical_radius and i != index:
            qml.PauliX(wires=i % num_qubits)  # Oracle flip approximation

    # Grover diffusion operator
    for wire in range(num_qubits):
        qml.Hadamard(wires=wire)

    if num_qubits > 1:
        qml.MultiControlledX(
            wires=list(range(num_qubits - 1)) + [num_qubits - 1],
            control_values=[1] * (num_qubits - 1),
        )

    for wire in range(num_qubits):
        qml.Hadamard(wires=wire)

    return qml.probs(wires=range(num_qubits))
```

```python
def estimate_all_local_densities():
    densities = []
    for i in range(num_points):
        probs = local_density_circuit(i)
        estimated_density = np.sum(probs)  # could sum over high-probability outcomes only
        densities.append(estimated_density)
    return np.array(densities)

densities = estimate_all_local_densities()
```

```python
@qml.qnode(dev)
def nearest_higher_circuit(index):
    bin_state = np.array(list(np.binary_repr(index, width=num_qubits)), dtype=int)

    qml.BasisState(bin_state, wires=range(num_qubits))
    for wire in range(num_qubits):
        qml.Hadamard(wires=wire)

    # Oracle: Flip for points with higher density and close distance
    for i in range(num_points):
        if densities[i] > densities[index] and dist_matrix[index, i] < critical_radius:
            qml.PauliX(wires=i % num_qubits)

    # Diffusion
    for wire in range(num_qubits):
        qml.Hadamard(wires=wire)

    if num_qubits > 1:
        qml.MultiControlledX(
            wires=list(range(num_qubits - 1)) + [num_qubits - 1],
            control_values=[1] * (num_qubits - 1)
        )

    for wire in range(num_qubits):
        qml.Hadamard(wires=wire)

    return qml.probs(wires=range(num_qubits))
```

```python
def find_nearest_higher_all():
    nearest_higher = []
    for i in range(num_points):
        probs = nearest_higher_circuit(i)
        nh = np.argmax(probs)
        nearest_higher.append(nh)
    return np.array(nearest_higher)

nearest_higher = find_nearest_higher_all()
```

```python
clusters = np.full(num_points, -1)
current_cluster = 0

for i in range(num_points):
    if nearest_higher[i] == i:  # Seed
        clusters[i] = current_cluster
        current_cluster += 1

for i in range(num_points):
    if clusters[i] == -1:
        neighbor = nearest_higher[i]
        while clusters[neighbor] == -1:
            neighbor = nearest_higher[neighbor]
        clusters[i] = clusters[neighbor]
```

```python
def distance_oracle(index, dist_matrix, critical_radius):
    """Returns a list of indices to mark (i.e., where dist(i, index) < dc)"""
    marked_indices = []
    for i in range(num_points):
        if dist_matrix[index, i] < critical_radius and i != index:
            marked_indices.append(i)
    return marked_indices
```

```python
def apply_oracle(marked_indices):
    """Apply phase flips to marked basis states directly in QNode scope."""
    for idx in marked_indices:
        bits = list(np.binary_repr(idx, width=num_qubits))

        # Apply PauliX to flip to |111...⟩ for target
        for i, b in enumerate(bits):
            if b == "0":
                qml.PauliX(wires=i)

        # Apply multi-controlled-Z (via MCX around Hadamard)
        qml.Hadamard(wires=num_qubits - 1)
        qml.MultiControlledX(
            control_wires=list(range(num_qubits - 1)),
            wires=num_qubits - 1,
            control_values="1" * (num_qubits - 1),
        )
        qml.Hadamard(wires=num_qubits - 1)

        # Undo the PauliX flips
        for i, b in enumerate(bits):
            if b == "0":
                qml.PauliX(wires=i)
```

```python
def diffusion_operator():
    """Inversion about the mean: H -> X -> MCX -> X -> H"""
    for wire in range(num_qubits):
        qml.Hadamard(wires=wire)
        qml.PauliX(wires=wire)

    qml.MultiControlledX(
        control_wires=list(range(num_qubits - 1)),
        wires=num_qubits - 1,
        control_values="1" * (num_qubits - 1),
    )

    for wire in range(num_qubits):
        qml.PauliX(wires=wire)
        qml.Hadamard(wires=wire)
```

```python
def grover_iterations(marked_indices, num_iter=1):
    @qml.qnode(dev)
    def circuit():
        # Initialize superposition
        for wire in range(num_qubits):
            qml.Hadamard(wires=wire)

        for _ in range(num_iter):
            apply_oracle(marked_indices)
            diffusion_operator()

        return qml.probs(wires=range(num_qubits))

    return circuit()
```

```python
index = 3  # base point to analyze
marked_indices = distance_oracle(index, dist_matrix, critical_radius)

# Determine number of Grover iterations: floor(pi/4 * sqrt(M))
num_iter = int(np.floor(np.pi / 4 * np.sqrt(len(marked_indices)))) or 1
grover_probs = grover_iterations(marked_indices, num_iter)

# Local density estimate: sum of amplified marked state probs
local_density_est = np.sum([grover_probs[i] for i in marked_indices])
print(f"Estimated local density for point {index}: {local_density_est}")
```

```python
def compute_local_densities(dist_matrix, critical_radius):
    """Compute Grover-estimated local density for all points."""
    densities = []
    for index in range(num_points):
        marked = distance_oracle(index, dist_matrix, critical_radius)

        if marked:
            num_iter = max(1, int(np.floor(np.pi / 4 * np.sqrt(len(marked)))))
            probs = grover_iterations(marked, num_iter)
            density = np.sum([probs[i] for i in marked])
        else:
            density = 0.0

        densities.append(density)

    return np.array(densities)
```

```python
def find_nearest_higher(index, densities, dist_matrix, d_max):
    """Grover-Enhanced Binary Search for Nearest Higher (Nj)"""
    rho_j = densities[index]
    dL = 0.0
    dT = d_max
    best_i = -1
    best_d = d_max

    for _ in range(num_qubits):  # log2 search rounds
        # 1. Build marked indices: those with rho_i > rho_j and dL < dist < dT
        marked = []
        for i in range(num_points):
            d = dist_matrix[index, i]
            if dL < d < dT and densities[i] > rho_j:
                marked.append(i)

        if not marked:
            dL = dT  # widen lower bound
            dT = (dT + best_d) / 2  # try midpoint
            continue

        num_iter = max(1, int(np.floor(np.pi / 4 * np.sqrt(len(marked)))))
        probs = grover_iterations(marked, num_iter)

        i_star = np.argmax([probs[i] for i in marked])
        candidate = marked[i_star]
        d_candidate = dist_matrix[index, candidate]

        # Found a candidate — tighten upper bound
        best_i = candidate
        best_d = d_candidate
        dT = (dL + d_candidate) / 2

    return best_i
```

```python
def compute_all_nearest_highers(densities, dist_matrix, d_max):
    nh = []
    for index in range(num_points):
        nj = find_nearest_higher(index, densities, dist_matrix, d_max)
        nh.append(nj if nj != -1 else index)  # default to self if none found
    return np.array(nh)
```

```python
def assign_clusters_from_nh(nearest_higher):
    clusters = np.full(num_points, -1)
    current_cluster = 0

    # Step 1: identify seeds (points that are their own nearest higher)
    for i in range(num_points):
        if nearest_higher[i] == i:
            clusters[i] = current_cluster
            current_cluster += 1

    # Step 2: assign followers by tracing nearest_higher links
    for i in range(num_points):
        if clusters[i] == -1:
            j = nearest_higher[i]
            path = [i]

            while clusters[j] == -1:
                path.append(j)
                j = nearest_higher[j]

            cluster_id = clusters[j]
            for node in path:
                clusters[node] = cluster_id

    return clusters
```

```python
import pennylane as qml
from pennylane import numpy as np
import matplotlib.pyplot as plt

from collections import Counter

#num_points = 20
critical_radius = 2
d_max = 1.5
rho_threshold = 0.1
distance_threshold = 1.0

num_qubits = int(np.ceil(np.log2(num_points)))
dev = qml.device("default.qubit", wires=num_qubits)

# Generate sample 3D points
np.random.seed(42)
points = np.random.rand(num_points, 3)
points += 0.05 * np.random.randn(*points.shape)
dist_matrix = np.linalg.norm(points[:, None, :] - points[None, :, :], axis=2)
```

```python
def distance_oracle(index, dist_matrix, radius):
    return [i for i in range(num_points) if i != index and dist_matrix[index, i] < radius]

def apply_oracle(marked_indices):
    for idx in marked_indices:
        bits = list(np.binary_repr(idx, width=num_qubits))
        
        for i, b in enumerate(bits):
            if b == "0":
                qml.PauliX(wires=i)
        
        # Apply controlled-Z on last qubit (target) with others as control
        qml.ctrl(qml.PauliZ, control=list(range(num_qubits - 1)))(wires=num_qubits - 1)
        
        # Undo the bit flips
        for i, b in enumerate(bits):
            if b == "0":
                qml.PauliX(wires=i)

def diffusion_operator():
    for wire in range(num_qubits):
        qml.Hadamard(wires=wire)
        qml.PauliX(wires=wire)

    qml.Hadamard(wires=num_qubits - 1)
    qml.MultiControlledX(
        wires=list(range(num_qubits - 1)) + [num_qubits - 1],
        control_values=[1] * (num_qubits - 1)
    )
    qml.Hadamard(wires=num_qubits - 1)

    for wire in range(num_qubits):
        qml.PauliX(wires=wire)
        qml.Hadamard(wires=wire)


def grover_iterations(marked_indices, num_iter=1):
    @qml.qnode(dev)
    def circuit():
        for wire in range(num_qubits):
            qml.Hadamard(wires=wire)

        for _ in range(num_iter):
            apply_oracle(marked_indices)
            diffusion_operator()

        return qml.probs(wires=range(num_qubits))

    return circuit()
```

```python
def compute_local_densities(dist_matrix, critical_radius):
    densities = []
    for index in range(num_points):
        marked = distance_oracle(index, dist_matrix, critical_radius)
        if marked:
            num_iter = max(1, int(np.floor(np.pi / 4 * np.sqrt(len(marked)))))
            probs = grover_iterations(marked, num_iter)
            #density = np.sum([probs[i] for i in marked])
            density = 0.5 * sum(probs[i] for i in marked)
            density += probs[index]
        else:
            density = 0.0
        densities.append(density)
    return np.array(densities)
```

```python
def find_nearest_higher(index, densities, dist_matrix, d_max):
    rho_j = densities[index]
    dL = 0.0
    dT = d_max
    best_i = -1
    best_d = d_max

    for _ in range(num_qubits):  # binary search steps
        marked = [i for i in range(num_points)
                  if i != index and dL < dist_matrix[index, i] < dT and densities[i] > rho_j]

        if not marked:
            dL = dT
            dT = (dT + best_d) / 2
            continue

        num_iter = max(1, int(np.floor(np.pi / 4 * np.sqrt(len(marked)))))
        probs = grover_iterations(marked, num_iter)
        scores = [probs[i] for i in marked]
        i_star = marked[np.argmax(scores)]

        best_i = i_star
        best_d = dist_matrix[index, best_i]
        dT = (dL + best_d) / 2

    return best_i if best_i != -1 else index


def compute_all_nearest_highers(densities, dist_matrix, d_max):
    return np.array([find_nearest_higher(i, densities, dist_matrix, d_max)
                     for i in range(num_points)])
```

```python
def assign_clusters_from_nh_with_outliers(nearest_higher, densities, dist_matrix,
                                          rho_threshold=0.1, distance_threshold=1.0):
    clusters = np.full(num_points, -1)
    current_cluster = 0

    # Step 1: identify seeds
    for i in range(num_points):
        if nearest_higher[i] == i:
            clusters[i] = current_cluster
            current_cluster += 1

    # Step 2: assign followers
    for i in range(num_points):
        if clusters[i] == -1:
            nj = nearest_higher[i]
            if densities[i] < rho_threshold and dist_matrix[i, nj] > distance_threshold:
                clusters[i] = -1  # Outlier
            else:
                path = [i]
                while clusters[nj] == -1 and nj != nearest_higher[nj]:
                    path.append(nj)
                    nj = nearest_higher[nj]
                cluster_id = clusters[nj] if clusters[nj] != -1 else -1
                for node in path:
                    clusters[node] = cluster_id

    return clusters
```

```python
densities = compute_local_densities(dist_matrix, critical_radius)
nearest_higher = compute_all_nearest_highers(densities, dist_matrix, d_max)
clusters = assign_clusters_from_nh_with_outliers(
    nearest_higher, densities, dist_matrix,
    rho_threshold=0.1, distance_threshold=1.0
)

# Count cluster occurrences
cluster_counts = Counter(clusters)

# Identify singleton clusters (size 1)
singleton_ids = [cid for cid, count in cluster_counts.items() if count == 1]

# Relabel singletons to -1
for i in range(len(clusters)):
    if clusters[i] in singleton_ids:
        clusters[i] = -1

print("Cluster labels with outliers:", clusters)
```
