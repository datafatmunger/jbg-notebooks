```python
import pennylane as qml
import numpy as np

dev = qml.device("default.qubit", wires=2)

@qml.qnode(dev)
def circuit1():
    qml.Hadamard(wires=0)
    qml.Hadamard(wires=1)
    qml.CNOT(wires=[0, 1])
    qml.Hadamard(wires=0)
    qml.Hadamard(wires=1)
    return qml.state()

@qml.qnode(dev)
def circuit2():
    qml.CNOT(wires=[1, 0])
    return qml.state()


qml.draw_mpl(circuit1)()
qml.draw_mpl(circuit2)()

# Extract the unitary matrix representations of the circuits
unitary1 = qml.matrix(circuit1)()
unitary2 = qml.matrix(circuit2)()

def test_circuit_equivalence():
    assert np.allclose(unitary1, unitary2), "Circuits are not equivalent!"
    print("Test passed: Circuits are equivalent.")

test_circuit_equivalence()
```

```python
@qml.qnode(dev)
def circuit3():
    qml.CZ(wires=[0, 1])
    return qml.state()

@qml.qnode(dev)
def circuit4():
    qml.CZ(wires=[1, 0])
    return qml.state()

qml.draw_mpl(circuit1)()
qml.draw_mpl(circuit2)()

# Extract the unitary matrix representations of the circuits
unitary3 = qml.matrix(circuit3)()
unitary4 = qml.matrix(circuit4)()

def test_circuit_equivalence():
    assert np.allclose(unitary3, unitary4), "Circuits are not equivalent!"
    print("Test passed: Circuits are equivalent.")

test_circuit_equivalence()
```

```python
@qml.qnode(dev)
def circuit5():
    qml.CNOT(wires=[1, 0])
    return qml.state()

@qml.qnode(dev)
def circuit6():
    qml.Hadamard(wires=1)
    qml.CZ(wires=[1, 0])
    qml.Hadamard(wires=1)
    return qml.state()

qml.draw_mpl(circuit5)()
qml.draw_mpl(circuit6)()

# Extract the unitary matrix representations of the circuits
unitary5 = qml.matrix(circuit5)()
unitary6 = qml.matrix(circuit6)()

def test_circuit_equivalence():
    assert np.allclose(unitary5, unitary6), "Circuits are not equivalent!"
    print("Test passed: Circuits are equivalent.")

test_circuit_equivalence()
```
