```python
import pennylane as qml
```

```python
dev = qml.device("default.qubit", wires=2)
```

```python
@qml.qnode(dev)
def circuit():
    qml.Hadamard(wires=[0])
    qml.CNOT(wires=[0, 1])
    return qml.probs()

qml.draw_mpl(circuit)()
```

```python
results = circuit()
results
```

```python
@qml.qnode(dev)
def circuit():
    qml.Hadamard(wires=[0])
    qml.Hadamard(wires=[1])
    qml.CNOT(wires=[1, 0])
    qml.Hadamard(wires=[0])
    qml.Hadamard(wires=[1])
    return qml.probs()
    
qml.draw_mpl(circuit)()
```
