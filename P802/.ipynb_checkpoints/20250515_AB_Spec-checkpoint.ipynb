```python
import numpy as np
from qutip import *

# Parameters
phi = np.pi / 2  # Example phase on red link
sites = 4
links = [(0, 1), (1, 2), (1, 3)]  # Blue links
red_link = (1, 3)

# Basic operators
a = destroy(2)
adag = a.dag()
n = adag * a
I = qeye(2)

def embed(op, site, N):
    """Place op at site in N-qubit tensor product"""
    ops = [I]*N
    ops[site] = op
    return tensor(ops)

def two_site_term(i, j, phase, N):
    """Construct a†_i t a_j + h.c. term"""
    a_i_dag = embed(adag, i, N)
    a_j = embed(a, j, N)
    term = a_i_dag * a_j * phase + (a_i_dag * a_j * phase).dag()
    return term

# Build Hamiltonian
H = 0
for i, j in links:
    H += two_site_term(i, j, phase=1.0, N=sites)

# Red link with Aharonov-Bohm phase
i, j = red_link
H += two_site_term(i, j, phase=np.exp(1j * phi), N=sites)

V = 1.0
for i in range(sites):
    for j in range(i+1, sites):
        ni = embed(n, i, sites)
        nj = embed(n, j, sites)
        H += V * ni * nj

# Diagonalize
eigvals = H.eigenenergies()
print(np.round(np.sort(eigvals), 6))
```

```python
import pennylane as qml
from pennylane.fermi import FermiC, FermiA, jordan_wigner
import networkx as nx
import numpy as np

# Parameters
phi = np.pi / 2 #0.5  # Aharonov-Bohm phase
eta = np.exp(1j * phi)  # Phase factor
N = 4  # Number of sites
edges = [(1, 2), (2, 3), (2, 4), (1, 3), (3, 4)]
red_link = (2, 3) # Edge where Aharonov-Bohm phase applies

# Build connectivity graph
G = nx.Graph()
G.add_edges_from(edges)

# Define path function between qubits
def jw_path(i, j):
    path = nx.shortest_path(G, i, j)
    return path[1:-1] if len(path) > 2 else []

# Get unique i<j with paths
nodes = sorted(set(i for e in edges for i in e))
#triples = [(i, j, jw_path(i, j)) for i in nodes for j in nodes if i < j]
triples = [(min(i,j), max(i,j), jw_path(i, j)) for (i, j) in edges]

# Anyonic Sigma dagger operator (fermionic version)
# def sigma_dagger(k, theta):
#     a_k_dag = FermiC(k)
#     a_k = FermiA(k)
#     n_k = a_k_dag * a_k
#     id_op = qml.fermi.FermiSentence({})
#     z_k = id_op - 2 * n_k
#     #term_I = 0.5 * np.exp(1 + 1j * theta) * id_op
#     #term_Z = 0.5 * np.exp(-1 + 1j * theta) * z_k
#     term_I = 0.5 * np.exp(1j * theta) * 1
#     term_Z = 0.5 * np.exp(1j * theta) * Z
#     return term_I + term_Z

def sigma_dagger(k, theta):
    id_op = qml.fermi.FermiSentence({})
    n_k = FermiC(k) * FermiA(k)
    z_k = id_op - 2 * n_k
    return 0.5 * (np.exp(1j * theta) + 1) * id_op + 0.5 * (np.exp(1j * theta) - 1) * z_k

# Build Hamiltonian using Jordan-Wigner transformation
coeffs = []
ops = []

for (i, j, path) in triples:
    i_idx = i - 1
    j_idx = j - 1
    a_i_dag = FermiC(i_idx)
    a_j = FermiA(j_idx)

    phase = eta if (i, j) == red_link or (j, i) == red_link else 1.0

    # Build Sigma dagger product along path
    sigmas = [sigma_dagger(k - 1, phi) for k in reversed(path)]

    f_op = a_i_dag
    for s in sigmas:
        f_op = f_op * s
    f_op = f_op * a_j
    f_op *= phase

    qubit_op = jordan_wigner(f_op)
    ops.extend([qubit_op, qubit_op.adjoint()])
    coeffs.extend([1.0, 1.0])

# Add interaction terms: n_i n_j
V = 1.0  # Interaction strength
for i in range(N):
    for j in range(i + 1, N):
        ni = FermiC(i) * FermiA(i)
        nj = FermiC(j) * FermiA(j)
        f_interaction = ni * nj
        qubit_op = jordan_wigner(f_interaction)
        ops.append(qubit_op)
        coeffs.append(V)

# Hamiltonian
H = qml.Hamiltonian(coeffs, ops)

# Iterate over each term in the Hamiltonian and print the Pauli string
for i, op in enumerate(H.ops):
    print(f"Pauli string of term {i+1}: {op}")
```

```python
import pennylane as qml
from pennylane import numpy as np
import networkx as nx

# PARAMETERS
phi = np.pi/2
N = 4 # Number of sites
edges = [(1, 2), (2, 3), (2, 4), (1, 3), (3, 4)]
red_link = (2, 3) # Edge where Aharonov-Bohm phase applies
AB_phase = np.exp(1j * phi)

# GRAPH FOR PATHS
G = nx.Graph()
G.add_edges_from(edges)
nodes = sorted(set(i for e in edges for i in e))

def jw_path(i, j):
    path = nx.shortest_path(G, i, j)
    return path[1:-1] if len(path) > 2 else []

# PAPER-DEFINED OPERATORS

def SP(k):  # σ₊
    return 0.5 * (qml.PauliX(k) - 1j * qml.PauliY(k))

def SM(k):  # σ₋
    return 0.5 * (qml.PauliX(k) + 1j * qml.PauliY(k))

def number_op(k):  # n_x = 0.5 * (1 + Z)
    return 0.5 * (qml.Identity(k) + qml.PauliZ(k))

# def sigma_anyonic(k, phi):
#     alpha = np.exp(1j * phi)
#     return 0.5 * (alpha + 1) * qml.Identity(k) + 0.5 * (alpha - 1) * qml.PauliZ(k)

def sigma_anyonic(k, phi):
    if np.isclose(phi % (2*np.pi), np.pi):  # fermion case
        return number_op(k) * (-2)
    else:
        alpha = np.exp(1j * phi)
        return 0.5 * (alpha + 1) * qml.Identity(k) + 0.5 * (alpha - 1) * qml.PauliZ(k)

def anyon_string(path, phi):
    if not path:
        return qml.Identity(wires=range(N))
    result = sigma_anyonic(path[0], phi)
    for k in path[1:]:
        result = result @ sigma_anyonic(k, phi)
    return result

#triples = [(i, j, jw_path(i, j)) for i in nodes for j in nodes if i < j]
triples = [(min(i,j), max(i,j), jw_path(i, j)) for (i, j) in edges]

ops = []
coeffs = []

for i, j, path in triples:
    i0, j0 = i - 1, j - 1
    path0 = [k - 1 for k in path]

    # Phase on red link only
    t = AB_phase if (i, j) == red_link or (j, i) == red_link else 1.0

    sigma_string = anyon_string(path0, phi)

    # a†_i σ a_j
    term1 = SP(i0) @ sigma_string @ SM(j0)
    term2 = term1.adjoint()

    ops.extend([term1, term2])
    coeffs.extend([t, np.conj(t)])

# ADD INTERACTION TERMS
V = 1.0
for i in range(N):
    for j in range(i + 1, N):
        ops.append(number_op(i) @ number_op(j))
        coeffs.append(V)

# BUILD HAMILTONIAN
H = qml.Hamiltonian(coeffs, ops)
Hmat = qml.matrix(H, wire_order=range(N))

# CHECK
print("Is H Hermitian?", np.allclose(Hmat, Hmat.conj().T))

#print(np.round(Hmat.copy(), decimals=2))
print(Hmat.shape)

# DIAGONALIZE
eigvals = np.linalg.eigvalsh(Hmat)
print("Eigenvalues from anyonic JW Hamiltonian:")
print(np.round(np.sort(eigvals), 6))
```

```python
import numpy as np
from scipy.linalg import expm
from itertools import combinations

# === Parameters ===
N = 4                   # Number of qubits (sites)
theta = 0.5             # Interaction strength
t = 1.0                 # Time evolution
initial_state = [0, 0, 1, 1]  # Initial state: |0011>

# === Step 1: Build 2-particle basis (binary strings with two 1's) ===
basis_2p = [np.array([int(i in comb) for i in range(N)]) for comb in combinations(range(N), 2)]
basis_states = [int("".join(map(str, b)), 2) for b in basis_2p]
basis_indices = {idx: i for i, idx in enumerate(basis_states)}
n_basis = len(basis_states)

# === Step 2: Build full 2^N x 2^N Hamiltonian for XX+YY terms ===
def pauli_term(op1, i, op2, j, N):
    """Build Kronecker product operator for op1_i * op2_j"""
    pauli = {'I': np.eye(2), 'X': np.array([[0, 1], [1, 0]]), 'Y': np.array([[0, -1j], [1j, 0]])}
    ops = [pauli['I']] * N
    ops[i] = pauli[op1]
    ops[j] = pauli[op2]
    result = ops[0]
    for op in ops[1:]:
        result = np.kron(result, op)
    return result

# Build full H
H_full = H_d.astype(complex)
ms_terms = 0
for i in range(N):
    for j in range(i+1, N):
        H_full += theta * (pauli_term('X', i, 'X', j, N) + pauli_term('Y', i, 'Y', j, N))
        ms_terms += 1  # Count each XX+YY term as 1 native MS-style gate

# === Step 3: Restrict to 2-particle subspace ===
H_2p = np.zeros((n_basis, n_basis), dtype=complex)
for i, bi in enumerate(basis_states):
    for j, bj in enumerate(basis_states):
        H_2p[i, j] = H_full[bi, bj]

# === Step 4: Time evolution ===
U = expm(-1j * t * H_2p)

# Prepare initial state vector in 2-particle basis
psi0 = np.zeros(n_basis, dtype=complex)
initial_index = basis_indices[int("".join(map(str, initial_state)), 2)]
psi0[initial_index] = 1.0
psi_t = U @ psi0

# === Step 5: Site occupation probabilities ===
site_occ = np.zeros(N)
for idx, amp in enumerate(psi_t):
    bin_state = basis_2p[idx]
    site_occ += np.abs(amp)**2 * bin_state

# === Final Output ===
print("Final state vector ψ(t):")
print(np.round(psi_t, 3))

print("\nSite occupation probabilities:")
for i in range(N):
    print(f"Site {i}: {site_occ[i]:.4f}")

print(f"\nGate count (MS-style XX+YY terms): {ms_terms}")
```

```python
def extract_ms_terms(Hmat, N, threshold=1e-8):
    """
    Extract XX and YY coefficients from Hmat for MS-compatible gate implementation.
    """
    from itertools import combinations

    # Define Pauli matrices
    I = np.eye(2)
    X = np.array([[0, 1], [1, 0]])
    Y = np.array([[0, -1j], [1j, 0]])

    def pauli_kron(op1, i, op2, j):
        ops = [I] * N
        ops[i] = op1
        ops[j] = op2
        out = ops[0]
        for k in range(1, N):
            out = np.kron(out, ops[k])
        return out

    ms_terms = {}
    count = 0

    for i, j in combinations(range(N), 2):
        Xij = pauli_kron(X, i, X, j)
        Yij = pauli_kron(Y, i, Y, j)

        # Project coefficient: Tr(P† H) / 2^N
        coeff_xx = np.trace(Xij.conj().T @ Hmat).real / 2**N
        coeff_yy = np.trace(Yij.conj().T @ Hmat).real / 2**N

        if abs(coeff_xx) > threshold or abs(coeff_yy) > threshold:
            ms_terms[(i, j)] = {'XX': coeff_xx, 'YY': coeff_yy}
            count += 1

    print(f"Detected {count} MS-compatible XX+YY terms:\n")
    for (i, j), terms in ms_terms.items():
        print(f"Pair ({i},{j}):  XX = {terms['XX']:.4f},  YY = {terms['YY']:.4f}")

    return ms_terms

H_full = H_d.astype(complex)
N = 4
ms_term_dict = extract_ms_terms(H_full, N)
```

```python
def extract_all_two_qubit_terms(Hmat, N, threshold=1e-8):
    """
    Extract all two-qubit Pauli-Pauli interaction coefficients from Hmat.
    Returns a dictionary {(i,j): {'XX': ..., 'XY': ..., ...}}.
    """
    from itertools import combinations, product

    # Define Pauli matrices
    I = np.eye(2)
    paulis = {
        'X': np.array([[0, 1], [1, 0]]),
        'Y': np.array([[0, -1j], [1j, 0]]),
        'Z': np.array([[1, 0], [0, -1]])
    }

    # Storage
    interaction_terms = {}
    count = 0

    for i, j in combinations(range(N), 2):
        pair_terms = {}
        for P, Q in product("XYZ", repeat=2):
            # Build operator P_i ⊗ Q_j
            ops = [I] * N
            ops[i] = paulis[P]
            ops[j] = paulis[Q]
            P_Q_op = ops[0]
            for k in range(1, N):
                P_Q_op = np.kron(P_Q_op, ops[k])

            # Project coefficient via Hilbert-Schmidt inner product
            coeff = np.trace(P_Q_op.conj().T @ Hmat) / (2**N)

            if abs(coeff) > threshold:
                pair_terms[f"{P}{Q}"] = coeff.real  # You can keep imag part if needed

        if pair_terms:
            interaction_terms[(i, j)] = pair_terms
            count += 1

    print(f"Detected {count} nonzero 2-qubit Pauli interaction pairs:\n")
    for (i, j), terms in interaction_terms.items():
        term_strs = [f"{k}={v:.4f}" for k, v in terms.items()]
        print(f"Pair ({i},{j}):  " + ",  ".join(term_strs))

    return interaction_terms

H_full = Hmat.astype(complex)
N = 4
ms_term_dict = extract_all_two_qubit_terms(H_full, N)
```

```python
def extract_all_pauli_terms(Hmat, N, threshold=1e-8, max_order=3):
    """
    Extract all single-, two-, and three-qubit Pauli terms from Hmat.
    Returns a dict: {term_string: coefficient}
    """
    from itertools import combinations, product

    paulis = {
        'I': np.eye(2),
        'X': np.array([[0, 1], [1, 0]]),
        'Y': np.array([[0, -1j], [1j, 0]]),
        'Z': np.array([[1, 0], [0, -1]])
    }

    results = {}

    # --- Loop over 1- to max_order-body terms ---
    for order in range(1, max_order + 1):
        for qubit_indices in combinations(range(N), order):
            for pauli_labels in product("XYZ", repeat=order):
                # Build full N-qubit operator
                ops = ['I'] * N
                for pos, q in enumerate(qubit_indices):
                    ops[q] = pauli_labels[pos]

                # Convert to matrix
                term_label = ''.join(f"{p}{i}" for i, p in enumerate(ops) if p != 'I')
                term_matrix = paulis[ops[0]]
                for op in ops[1:]:
                    term_matrix = np.kron(term_matrix, paulis[op])

                # Project coefficient
                coeff = np.trace(term_matrix.conj().T @ Hmat) / (2 ** N)
                if abs(coeff) > threshold:
                    results[term_label] = coeff.real  # or complex if needed

    # --- Print summary ---
    print(f"Extracted {len(results)} Pauli terms (up to {max_order}-body):\n")
    for term, coeff in sorted(results.items()):
        print(f"{term:10}: {coeff:+.4f}")

    return results

#H_full = Hmat.astype(complex)
pauli_terms = extract_all_pauli_terms(Hmat, N=4, max_order=3)
```
